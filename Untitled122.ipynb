Untuk mengintegrasikan model Hugging Face, kita perlu menginstal pustaka `transformers` dan kemudian memodifikasi `app.py` untuk menggunakan `pipeline` dari `transformers`. Kita akan menggunakan model `microsoft/DialoGPT-medium` sebagai contoh untuk percakapan.

Sekarang, kita akan memodifikasi `app.py` untuk menggunakan model Hugging Face. Perhatikan bahwa konfigurasi Gemini API akan dihapus, dan gantinya kita akan menginisialisasi Hugging Face `conversational` pipeline.

```python
%%writefile app.py

import streamlit as st
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch

st.set_page_config(
    page_title="Chatbot Pendidikan",
    page_icon="🎓",
    layout="centered",
    initial_sidebar_state="auto"
)

st.title("🎓 Chatbot Pendidikan")

# Inisialisasi st.session_state.messages terlebih dahulu
if "messages" not in st.session_state:
    st.session_state.messages = []

# Muat tokenizer dan model DialoGPT secara langsung
if "tokenizer" not in st.session_state:
    st.session_state.tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
if "model" not in st.session_state:
    st.session_state.model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

# Inisialisasi riwayat percakapan untuk model Hugging Face
if "chat_history_ids" not in st.session_state:
    st.session_state.chat_history_ids = None

# Tampilkan pesan dari riwayat chat saat aplikasi dimulai
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

# Tangani input pengguna
if prompt := st.chat_input("Ajukan pertanyaan Anda tentang pendidikan..."):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    with st.chat_message("assistant"):
        message_placeholder = st.empty()
        full_response = ""
        try:
            # Encode input pengguna dan riwayat percakapan
            new_user_input_ids = st.session_state.tokenizer.encode(prompt + st.session_state.tokenizer.eos_token, return_tensors='pt')

            # Gabungkan input pengguna baru dengan riwayat percakapan sebelumnya (jika ada)
            bot_input_ids = torch.cat([st.session_state.chat_history_ids, new_user_input_ids], dim=-1) if st.session_state.chat_history_ids is not None else new_user_input_ids

            # Hasilkan respons
            st.session_state.chat_history_ids = st.session_state.model.generate(
                bot_input_ids,
                max_length=1000,
                pad_token_id=st.session_state.tokenizer.eos_token_id,
                no_repeat_ngram_size=3,
                do_sample=True,
                top_k=100,
                top_p=0.7,
                temperature=0.8
            )

            # Dekode respons bot
            full_response = st.session_state.tokenizer.decode(st.session_state.chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

        except Exception as e:
            full_response = f"Maaf, terjadi kesalahan saat menghubungi model Hugging Face: {e}"

        message_placeholder.markdown(full_response)
    st.session_state.messages.append({"role": "assistant", "content": full_response})
```

**Reasoning**:
The embeddings for the knowledge base have been created. To enable efficient retrieval of relevant documents, the next step is to create a FAISS index from these embeddings. This will allow for fast similarity searches.



```python
import faiss
import numpy as np

print('Membuat FAISS index dari embeddings...')
dimension = knowledge_base_embeddings.shape[1] # Dimensi embeddings
faiss_index = faiss.IndexFlatL2(dimension) # Menggunakan indeks Euclidean distance (L2)
faiss_index.add(np.array(knowledge_base_embeddings).astype('float32')) # Menambahkan embeddings ke indeks
print(f'FAISS index berhasil dibuat dengan {faiss_index.ntotal} dokumen.')
```

**Reasoning**:
I will define the `retrieve_documents` function as specified in the instructions, which will convert a query into an embedding, use the FAISS index to find the top `k` most similar documents, and then return the original text of these documents from the knowledge base. After defining, I will test it with a sample query.



```python
def retrieve_documents(query, embedding_model, faiss_index, knowledge_base, k=3):
    # 1. Ubah query pengguna menjadi embedding
    query_embedding = embedding_model.encode([query])

    # 2. Gunakan faiss_index untuk mencari k dokumen teratas
    distances, indices = faiss_index.search(np.array(query_embedding).astype('float32'), k)

    # 3. Ambil teks asli dari knowledge_base
    retrieved_docs = [knowledge_base[idx] for idx in indices[0]]

    return retrieved_docs

print('Fungsi retrieve_documents berhasil didefinisikan.')

# Uji fungsi retrieval dengan sebuah contoh query
example_query = "Apa manfaat utama dari pendidikan?"
print(f"\nMengambil dokumen yang relevan untuk query: '{example_query}'")
relevant_documents = retrieve_documents(example_query, embedding_model, faiss_index, knowledge_base, k=2)

print("Dokumen yang diambil:")
for i, doc in enumerate(relevant_documents):
    print(f"[{i+1}] {doc}")
```

```python
%%writefile app.py

import streamlit as st
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline
import torch # Meskipun tidak digunakan secara langsung di sini, bisa berguna untuk manajemen device jika model lebih besar

st.set_page_config(
    page_title="Chatbot Pendidikan (RAG)",
    page_icon="🎓",
    layout="centered",
    initial_sidebar_state="auto"
)

st.title("🎓 Chatbot Pendidikan (RAG)")

# 1. Knowledge Base
knowledge_base = [
    "Pendidikan adalah proses memfasilitasi pembelajaran, atau pemerolehan pengetahuan, keterampilan, nilai-nilai, moral, kepercayaan, dan kebiasaan. Metode pendidikan meliputi pengajaran, pelatihan, penceritaan, diskusi, dan penelitian terarah.",
    "Sistem pendidikan formal biasanya terbagi menjadi beberapa tingkatan: prasekolah, sekolah dasar, sekolah menengah, dan pendidikan tinggi.",
    "Sekolah dasar (atau sekolah dasar) adalah tahap pertama pendidikan formal. Biasanya berlangsung dari sekitar usia 6 hingga 12 tahun.",
    "Sekolah menengah mengikuti sekolah dasar dan mempersiapkan siswa untuk pendidikan tinggi atau masuk ke angkatan kerja. Umumnya diperuntukkan bagi remaja antara usia 12 dan 18 tahun.",
    "Pendidikan tinggi, juga disebut pendidikan tersier, adalah pendidikan tingkat ketiga yang mengarah pada perolehan gelar akademik atau diploma profesional. Ini mencakup universitas, perguruan tinggi, dan institusi teknis.",
    "Kurikulum adalah seperangkat kursus dan kontennya yang ditawarkan di sekolah atau universitas. Ini mencakup mata pelajaran yang diajarkan, penilaian, dan bahan ajar.",
    "Metode pengajaran adalah prinsip dan metode yang digunakan untuk instruksi oleh guru. Metode ini bervariasi tergantung pada materi pelajaran dan kebutuhan siswa.",
    "Pembelajaran daring (e-learning) adalah pendidikan yang disampaikan melalui Internet, menggunakan perangkat digital seperti komputer, tablet, atau smartphone.",
    "Manfaat pendidikan meliputi peningkatan peluang kerja, pendapatan yang lebih tinggi, pengembangan pribadi, dan kontribusi pada masyarakat.",
    "Tantangan dalam pendidikan meliputi akses yang tidak setara, kualitas yang tidak memadai, kurangnya sumber daya, dan kebutuhan untuk beradaptasi dengan perubahan teknologi."
]

# Inisialisasi model embedding dan FAISS index di session state
# Ini akan memastikan model hanya dimuat sekali
if "embedding_model" not in st.session_state:
    st.session_state.embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
    st.session_state.knowledge_base_embeddings = st.session_state.embedding_model.encode(knowledge_base, show_progress_bar=False)
    dimension = st.session_state.knowledge_base_embeddings.shape[1]
    st.session_state.faiss_index = faiss.IndexFlatL2(dimension)
    st.session_state.faiss_index.add(np.array(st.session_state.knowledge_base_embeddings).astype('float32'))

# Inisialisasi model generatif (Question Answering) di session state
if "qa_pipeline" not in st.session_state:
    st.session_state.qa_pipeline = pipeline('question-answering', model='distilbert-base-cased-distilled-squad')

# Fungsi Retrieval
def retrieve_documents(query, k=3):
    query_embedding = st.session_state.embedding_model.encode([query])
    distances, indices = st.session_state.faiss_index.search(np.array(query_embedding).astype('float32'), k)
    retrieved_docs = [knowledge_base[idx] for idx in indices[0]]
    return retrieved_docs

# Inisialisasi riwayat pesan Streamlit
if "messages" not in st.session_state:
    st.session_state.messages = []

# Tampilkan pesan dari riwayat chat saat aplikasi dimulai
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

# Tangani input pengguna
if prompt := st.chat_input("Ajukan pertanyaan Anda tentang pendidikan..."):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    with st.chat_message("assistant"):
        message_placeholder = st.empty()
        full_response = ""
        try:
            # 1. Retrieve relevant documents
            retrieved_docs = retrieve_documents(prompt, k=3)
            context = " ".join(retrieved_docs)

            # 2. Generate answer using QA pipeline
            # Pastikan pipeline menerima input dalam format yang benar
            answer = st.session_state.qa_pipeline(question=prompt, context=context)
            full_response = answer['answer']

        except Exception as e:
            full_response = f"Maaf, terjadi kesalahan saat memproses pertanyaan Anda: {e}"

        message_placeholder.markdown(full_response)
    st.session_state.messages.append({"role": "assistant", "content": full_response})
```
